```python
import matplotlib.pyplot as plt

# Input data
cluster_values = {
    0: 514.1529411764706,
    1: 28635.298039215675,
    2: 2432132.811764714,
    3: 51466.294117647056
}

# Preparing data for plotting
clusters = list(cluster_values.keys())
values = list(cluster_values.values())

# Creating the bar plot
plt.figure(figsize=(10, 6))
plt.bar(clusters, values, color='skyblue')
plt.xlabel('Cluster')
plt.ylabel('Value')
plt.title('Cluster Values')
plt.xticks(clusters)
plt.yscale('log')  # Using a logarithmic scale due to large value differences
plt.show()
```

```python
total = sum(values)
percentages = [(value / total) * 100 for value in values]

# Creating the bar plot with percentages
print(percentages)
plt.figure(figsize=(10, 6))
plt.bar(clusters, percentages, color='lightgreen')
plt.xlabel('Cluster')
plt.ylabel('Percentage (%)')
plt.title('Cluster Values as Percentages')
plt.xticks(clusters)
plt.show()
```
