```python
import os
from osgeo import gdal
import numpy as np

# Define input and output directories
input_folder_path = "C:\\Users\\Musae\\Documents\\GitHub-REPOs\\Senior-project-main\\NDVI-Data"
output_folder_path = "G:\\My Drive\\NumPyArrays"

# Create the output directory if it doesn't exist
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# Iterate through each .tif file in the input directory
for filename in os.listdir(input_folder_path):
    if filename.endswith('.tif'):
        # Construct the full path to the .tif file
        file_path = os.path.join(input_folder_path, filename)
        
        # Open the dataset
        dataset = gdal.Open(file_path)
        
        # Proceed if the dataset could be opened
        if dataset:
            # Read the single band's data as a NumPy array
            array = dataset.GetRasterBand(1).ReadAsArray()
            
            # Define the output .npy file path
            output_file_path = os.path.join(output_folder_path, f'{os.path.splitext(filename)[0]}.npy')
            
            # Save the array as a .npy file
            np.save(output_file_path, array)
            
            print(f"Saved NumPy array for {filename} to {output_file_path}")
        else:
            print(f"Failed to open file {filename}")
```
