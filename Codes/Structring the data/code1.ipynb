```python
import pandas as pd

def load_climate_data(filepath):
    climate_df = pd.read_csv(filepath)
    # Splitting the 'Year/Month' column and expanding into separate columns
    split_columns = climate_df['Year/Month'].str.split('/', expand=True)
    climate_df['Year'] = split_columns[0].astype(int)
    climate_df['Month'] = split_columns[1].astype(int)
    return climate_df

# Update the path to your CSV file
climate_data = load_climate_data("C:\\Users\\Musae\\Documents\\GitHub-REPOs\\Senior-project_Doc\\monthly_averages_formatted.csv")
```
