```python
import pandas as pd

# Replace 'your_file.csv' with the path to your actual CSV file
file_path = "C:\\Users\\Musae\\Documents\\GitHub-REPOs\\Senior-project_Doc\\Docs\\Riyadh_Daily_2018_12_01_2023_12_31.xlsx"

# Read the .xlsx file into a DataFrame
df = pd.read_excel(file_path)
```

```python
import pandas as pd
import numpy as np

# Assuming 'df' is your DataFrame

# Combine YEAR, MO, DY into a date string column
df['DATE_STR'] = df['YEAR'].astype(str) + '-' + df['MO'].astype(str).str.zfill(2) + '-' + df['DY'].astype(str).str.zfill(2)

# Convert the date string column to datetime
df['DATE'] = pd.to_datetime(df['DATE_STR'], format='%Y-%m-%d')

# Set DATE as the index
df.set_index('DATE', inplace=True)

# Exclude non-numeric columns before resampling and aggregating
numeric_df = df.select_dtypes(include=[np.number])

# Group by year and month, then calculate the mean for numeric columns
monthly_averages = numeric_df.resample('M').mean()

# Prepare data for the new DataFrame
data = []
for date, row in monthly_averages.iterrows():
    year_month = date.strftime('%Y/%m')
    temp_avg = (row.get('Temp_Max', 0) + row.get('Temp_Min', 0)) / 2
    prectotcorr_avg = row.get('PRECTOTCORR', 0)
    data.append([year_month, temp_avg, prectotcorr_avg])

# Create a new DataFrame
monthly_averages_df = pd.DataFrame(data, columns=['Year/Month', 'Temp Average', 'PRECTOTCORR Average'])

# Save the new DataFrame to a CSV file
csv_file_path = 'monthly_averages_formatted.csv'
monthly_averages_df.to_csv(csv_file_path, index=False)

print(f'Monthly averages saved to {csv_file_path}')
```
